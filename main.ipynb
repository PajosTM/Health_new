```python
import pandas as pd
import numpy as np
```

```python
nutritional_data = pd.read_excel('NutritionalFacts.xlsx' )
```

```python
nutritional_data.head()
```

```python
nutritional_data = nutritional_data.drop(0, axis = 0)
```

```python
nutritional_data.head()
```

```python
nutritional_data['Name'] = nutritional_data['Food and Serving'].apply(lambda x: x.split(',')[0])
nutritional_data['Category'] = nutritional_data['Food Type'].apply(lambda x: x.split(',')[0])
```

```python
nutritional_data.sample(5)
```

```python
main_df = nutritional_data.drop(['Food and Serving', 'Food Type'], axis = 1)
```

```python
main_df.sample(5)
```

```python
np.random.seed(42)  # For reproducibility

num_users = 100

# Sample user profile data
user_profiles = pd.DataFrame({
    'UserID': range(1, num_users + 1),
    'Age': np.random.randint(18, 65, size=num_users),
    'Gender': np.random.choice(['Male', 'Female'], size=num_users),
    'Dietary Preferences': np.random.choice(['Vegetarian', 'Vegan', 'Omnivore'], size=num_users),
    'Favorite Foods': [np.random.choice(main_df['Name'], size=3, replace=False).tolist() for _ in range(num_users)],
    'Disliked Foods': [np.random.choice(main_df['Name'], size=1, replace=False).tolist() for _ in range(num_users)],
    'Allergies': np.random.choice(['None', 'Nuts', 'Dairy', 'Gluten'], size=num_users),
    'Intolerances': np.random.choice(['None', 'Lactose', 'Gluten'], size=num_users),
    'Health Objectives': np.random.choice(['Weight Loss', 'Muscle Gain', 'General Fitness'], size=num_users),
    'Activity Level': np.random.choice(['Sedentary', 'Moderately active', 'Very active'], size=num_users),
    'Meals per Day': np.random.randint(2, 6, size=num_users)
})

user_profiles.head()
user_profiles.to_csv('user_profile.csv')
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
```

```python
# Preprocess the nutritional data
features = main_df.drop(columns=['Name', 'Category'])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
scaled_features[:5]
```

```python
def get_preferred_foods(user_profile, main_df):
    preferred_foods = main_df[~main_df['Name'].isin(user_profile['Disliked Foods'])]
    return preferred_foods
```

```python
# Train KNN model
knn = NearestNeighbors(n_neighbors=5, metric='euclidean')
knn.fit(scaled_features)
```

```python
# Function to recommend foods for a user
def recommend_foods(user_profile, main_df, knn_model, scaler):
    preferred_foods = get_preferred_foods(user_profile, main_df)
    preferred_features = preferred_foods.drop(columns=['Name', 'Category'])
    
    if preferred_features.empty:
        return []
    
    scaled_preferred_features = scaler.transform(preferred_features)
    
    # Example: using the average nutritional profile of the preferred foods as the query point
    query_point = np.mean(scaled_preferred_features, axis=0).reshape(1, -1)
    
    distances, indices = knn_model.kneighbors(query_point)
    
    recommended_foods = preferred_foods.iloc[indices[0]]
    
    return recommended_foods
```

```python
# Example recommendation for the first user profile
user_profile_example = user_profiles.iloc[0]
```

```python
recommendations = recommend_foods(user_profile_example, main_df, knn, scaler)
recommendations['Name'].values
```
